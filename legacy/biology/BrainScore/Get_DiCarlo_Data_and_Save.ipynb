```python
import brainscore_vision
import pandas as pd
import numpy as np
#import matplotlib.pyplot as plt

neural_data = brainscore_vision.load_dataset("MajajHong2015.public")
neural_data = neural_data.transpose('presentation', 'neuroid', 'time_bin')
neural_data
```

```python
neural_data['neuroid_id'].to_numpy()
```

```python
#region_name='V4'
region_name='IT'
animal_name='Tito'
#animal_name='Chabo'
 

num_reps=np.max(neural_data['repetition'].to_numpy())
num_reps=28

aa=np.ones(5000)
for i in range(num_reps):
    benchmark_data = neural_data.sel(region=region_name,animal=animal_name,repetition=i)#Tito#Chabo
    stim=benchmark_data['stimulus'].to_numpy()
    
    ustim,idcs=np.unique(stim, return_index=True)
    ac=np.zeros(5000)
    ac[ustim]=1
    aa=aa*ac
    
uniques=np.where(aa)
    

Xs=[]
stims=[]
objs=[]
for i in range(num_reps):
    benchmark_data = neural_data.sel(region=region_name,animal=animal_name,repetition=i)#Tito#Chabo
    stim=benchmark_data['stimulus'].to_numpy()
    #print(np.shape(stim))
    X=benchmark_data.squeeze('time_bin') 
    obj=benchmark_data['object_name'].to_numpy()
    #print(np.shape(X))

    ustim,idcs=np.unique(stim, return_index=True)
    
    Xun=X[idcs,:]
    objun=obj[idcs]
    
    uniques, idx1, idx2=np.intersect1d(ustim, uniques, return_indices=True)
    
    ustim=ustim[idx1]
    Xun=Xun[idx1,:]
    objun=objun[idx1]

    stims.append(ustim)
    Xs.append(Xun)
    objs.append(objun)
stims=np.array(stims)
Xs=np.array(Xs)
objs=np.array(objs)


np.savez(region_name+'_'+animal_name+'.npz',stims=stims,Xs=Xs,objs=objs)

```

```python
unobj=np.unique(objs[0])
Xos=[]
for objn in unobj:
    obj_idx=objs[0]==objn
    Xos.append(Xs[:,obj_idx,:])
    

```
